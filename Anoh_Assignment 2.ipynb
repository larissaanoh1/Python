```python
# Create Increasing plant growth dataframe
dataframe = pd.DataFrame({'plant_food_A': np.repeat(['grow_light1', 'grow_light2'], 6),
                          'plant_food_B': np.repeat(['grow_light1', 'grow_light2'], 6),
                          'value': [9.2, 9.4, 8.9, 7.1, 7.2, 8.5, 8.5, 9.2, 8.9,  5.5, 5.8, 7.6]})
# Performing two-way ANOVA
model = ols('value ~ C(plant_food_A) + C(plant_food_B) +\
C(plant_food_A):C(plant_food_B)', 
            data=dataframe).fit()
result = sm.stats.anova_lm(model, type=2)
    
# Print the result
print(result)
```

```python
def personChiSqGof(df1, Year, exp=None):
    myFreq = df1[Year].value_counts()
    df = len(myFreq) - 1

    if exp == None:
        minE = sum(myFreq)/len(myFreq)
        chiVal, pval = chisquare(myFreq)
    else:
        minE = min(exp)
        chiVal, pval = chisquare(myFreq, exp)
    warning = None
    if minE < 5: 
        warning = 'minimum expected count less than 5, chi-square test results not reliable'
        
    return chiVal, pval, df, minE, warning 
```

```python
personChiSqGof(df1, 'Year')
```
