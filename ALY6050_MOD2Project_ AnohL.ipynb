```python
import random
# Benefits and costs for Dam #1
d1_benefits = np.array([
    [1.1, 2, 2.8],   # improved navigation
    [8, 12, 14.9],   # hydroelectric power
    [1.4, 1.4, 2.2], # fish and wildlife
    [6.5, 9.8, 14.6],   # recreation
    [1.7, 2.4, 3.6],    # flood control
    [0, 1.6, 2.4]])    # commercial development
d1_costs = np.array([[13.2, 14.2, 19.1], 
            [3.5, 4.9, 7.4]])
d1_years = 30

# Benefits and costs for Dam #2
d2_benefits = np.array([
    [2.1, 3, 4.8],   # improved navigation
    [8.7, 12.2, 13.6],   # hydroelectric power
    [2.3, 3, 3], # fish and wildlife
    [5.9, 8.7, 15],   # recreation
    [0, 3.4, 3.4],    # flood control
    [0, 1.2, 1.8]])    # commercial development
d2_costs = np.array([[12.8, 15.8, 20.1], 
            [3.8, 5.7, 8]])
d2_years = 30

# Simulation of 10,000 benefit-cost ratios for Dam #1 and Dam #2
alpha_1 = np.sum(d1_benefits, axis=1) / np.sum(d1_costs)
alpha_2 = np.sum(d2_benefits, axis=1) / np.sum(d2_costs)
alpha_1_sim = np.random.normal(np.mean(alpha_1), np.std(alpha_1), 10000)
alpha_2_sim = np.random.normal(np.mean(alpha_2), np.std(alpha_2), 10000)

# Print results
print("Benefit-cost ratios for Dam #1:", alpha_1)
print("Benefit-cost ratios for Dam #2:", alpha_2)
```

```python
from scipy.stats import skew, norm


def proportion_check(alpha_list, x_i):
  prop = sum([1 for x in alpha_list if x > x_i]) / len(alpha_list)
  return prop


def get_metrics(alpha_list):
  minimum = min(alpha_list)
  maximum = max(alpha_list)
  mean = np.mean(alpha_list)
  median = np.median(alpha_list)
  var = np.var(alpha_list)
  std = np.std(alpha_list)
  skewness = sum([((x - mean) / var) ** 3 for x in alpha_list]) * len(alpha_list) / ((len(alpha_list) - 1) * (len(alpha_list) - 2))
  prop_2 = proportion_check(alpha_list, 2)
  prop_1_8 = proportion_check(alpha_list, 1.8)
  prop_1_5 = proportion_check(alpha_list, 1.5)
  prop_1_2 = proportion_check(alpha_list, 1.2)
  prop_1 = proportion_check(alpha_list, 1)

  return [minimum, maximum, mean, median, var, std, skewness, prop_2, prop_1_8, prop_1_5, prop_1_2, prop_1]


df = {"Distributions": ["Minimum", "Maximum", "Mean", "Median", "Variance", "Standard Deviation", "SKEWNESS", "P(𝛂i > 2)", "P(𝛂i > 1.8)", "P(𝛂i > 1.5)", "P(𝛂i > 1.2)", "P(𝛂i > 1)"],
      "alpha_1": get_metrics(alpha_1),"alpha_2": get_metrics(alpha_2)}
df = pd.DataFrame(df)
df
```
